La siguiente es la instrucción que hace la nube de palabras del "texto".

```python
#Define la funcion plot para archivo texto
def makeImage(texto, nomarchi):
#    wc = WordCloud(background_color="white", max_words=1000)
    wc = WordCloud(width=3000, 
                    height=2000,
                    background_color="white", 
                    mode="RGBA",max_words=50,
                    stopwords=stop)
    # generate word cloud
    wc.generate(texto)
    
    # set the word color to navy
    wc.recolor(color_func = navy_color_func)

    # show
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.savefig(nomarchi+'.png',dpi=300) #cambiar nombre de figura
    plt.show()
```

Ahora el ejemplo de la nube de palabras

```python
makeImage(text,nomarchi='nube')
```
